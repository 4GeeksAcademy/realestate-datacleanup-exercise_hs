#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
import seaborn as sns


# List of populations of interest
populations_of_interest = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Normalize the prices for each population
normalized_subset_ds = df_cinturonsur.copy()
for population in populations_of_interest:
    subset_population = normalized_subset_ds[normalized_subset_ds['level5'] == population]
    max_price = subset_population['price'].max()
    min_price = subset_population['price'].min()
    normalized_subset_ds.loc[subset_population.index, 'normalized_price'] = (subset_population['price'] - min_price) / (max_price - min_price)

# Create a single histogram plot for normalized prices using Seaborn
plt.figure(figsize=(10, 6))
for population in populations_of_interest:
    subset_population = normalized_subset_ds[normalized_subset_ds['level5'] == population]
    sns.histplot(data=subset_population, x='normalized_price', bins=20, kde=True, label=population)

plt.title('Normalized Prices Histogram for Different Populations')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆)

```python


# List of populations to compare
populations_to_compare = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

# Create scatter plots for each population
for i, population in enumerate(populations_to_compare):
    subset_population = df_fotocasa_limpio[df_fotocasa_limpio['level5'] == population]
    sns.scatterplot(data=subset_population, x='surface', y='price', ax=axes[i])
    axes[i].set_title(f'Scatter Plot of Surface vs. Price - {population}')
    axes[i].set_xlabel('Surface Area')
    axes[i].set_ylabel('Price')

# Adjust layout
plt.tight_layout()
plt.show()
```
